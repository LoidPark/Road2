```python
def remove_currency(val):
    out=val.replace('€','')
    if 'M' in out:
        out= float(out.replace('M', ''))*100000
    elif 'K' in out:
        out=float(out.replace('K', ''))*1000
    return out
```

```python
fifa_new = fifa.copy()
```

```python
fifa_new['Wage']=fifa_new['Wage'].apply(lambda x:remove_currency(x))
fifa_new['Value']=fifa_new['Value'].apply(lambda x:remove_currency(x))
```

```python
fifa_new['Wage']=pd.to_numeric(fifa_new['Wage'], errors='coerce')
fifa_new['Wage']
```
